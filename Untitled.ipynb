```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dset=pd.read_csv("houses.csv")
dset.head()
```

```python
X=dset.drop("price",axis=1)
y=dset.loc[:,"price"]
y=np.array(y).reshape(-1,1)
y[:10]
#drop missing data
dset.dropna()
dset.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44, shuffle =True)
```

```python
#scalling_standization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
print("mean",scaler.mean_)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
X_train_scaled.describe()
print("X_test_scaled",X_test_scaled.head())


print("X_train_scaled",X_train_scaled.head())


#############
scalery=StandardScaler()

scalery.fit(y_train)
y_train_scaled=scalery.transform(y_train)
y_test_scaled=scalery.transform(y_test)
print("mean",scalery.mean_)

y_test_scaled =np.array(y_test_scaled[:5])

print("y_test_scaled",y_test_scaled)

y_train_scaled =  np.array(y_train_scaled[:5])

print("y_train_scaled",y_train_scaled)
#############
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['bedrooms'], X_train['sqft_living'],c=y_train)
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['bedrooms'], X_train_scaled['sqft_living'],c=y_train)
ax2.set_title("After Scaling")
plt.show()
```

```python
#minmax_or_normlization
from sklearn.preprocessing import MinMaxScaler
scalerminmax=MinMaxScaler()
scalerminmax.fit(X_train)
X_train_scaledMinmax=scalerminmax.transform(X_train)
X_test_scaledMinmax=scalerminmax.transform(X_test)
######
X_test_scaledMinmax=pd.DataFrame(X_test_scaledMinmax,columns=X_test.columns)
X_train_scaledMinmax=pd.DataFrame(X_train_scaledMinmax,columns=X_test.columns)
#########

X_train_scaledMinmax.describe()
##########
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['bedrooms'], X_train['sqft_living'],c=y_train)
ax1.set_title("Before normalization")
ax2.scatter(X_train_scaledMinmax['bedrooms'], X_train_scaledMinmax['sqft_living'],c=y_train)
ax2.set_title("After normalization")
plt.show()
```
